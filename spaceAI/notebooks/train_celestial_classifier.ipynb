# Celestial Object Classification
Training a Decision Tree classifier on celestial data

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib
```

## Load dataset

```python
data = pd.read_csv('../data/celestial_objects.csv')
X = data[['orbital_period', 'axial_tilt', 'mass']]
y = data['type']

print("Dataset shape:", X.shape)
print("Classes:", y.unique())
```

## Split and train

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

print("Model trained")
```

## Evaluate

```python
predictions = model.predict(X_test)
print('Accuracy:', accuracy_score(y_test, predictions))
```

## Save model

```python
joblib.dump(model, '../models/celestial_classifier.pkl')
print('Model saved to models/celestial_classifier.pkl')
```
